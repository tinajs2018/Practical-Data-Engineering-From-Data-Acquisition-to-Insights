```python
#Read the CSV Data
data=pd.read_csv("Unemployment rate among black Americans.csv")
data.head()
```

### Perform Data Manipulation

```python
# Filter the data for a specific period
product_data = data[data['Period'] == 'M01']
product_data 
#all the values of january
```

```python
# Group the data by date and calculate the total sales for each date
daily_sales = data.groupby('Year')['Value'].sum()
daily_sales.head()
                                   
```

```python
#Save the Processed Data
daily_sales.to_csv("daily_salesby year sum",index=False)
```
